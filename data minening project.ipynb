```python
#gerekli kütüphanelerin import edilmesi
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#adresi verilien csv formatlı veri setinin okunması
#head() fonksiyonu ilk 5 satırı listelenmesini sağlar, aldığı parametre ile istediği sayıda satır yazdırabilir
veri = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
veri.head()
```

```python
#seçtiğimiz verilerin datasetten düşürülmesini sağlar
veri.drop(["EmployeeNumber","EmployeeCount","StandardHours","Over18"],axis=1,inplace=True)
```

```python
#uniqe fonksiyonu listedeki farklı değerleri bulur
categorical_col = []
for column in veri.columns:
    if veri[column].dtype == object and len(veri[column].unique()) <= 50:
        categorical_col.append(column)
        print(f"{column} : {veri[column].unique()}")
        print("====================================")
```

```python
#astype() method is used to cast a pandas object to a specified dtype. 
#astype() function also provides the capability to convert any suitable existing 
#column to categorical type. DataFrame. astype() function comes very handy when we 
#want to case a particular column data type to another data type.
veri['Attrition'] = veri.Attrition.astype("category").cat.codes
```

```python
#veri görselleştirme
veri.Attrition.value_counts()
```

```python
# Visulazing the distibution of the data for every feature
veri.hist(edgecolor='black', linewidth=1.2, figsize=(20, 20));
```

```python
# Plotting how every feature correlate with the "target"
sns.set(font_scale=1.2)
plt.figure(figsize=(20, 10))

for i, column in enumerate(categorical_col, 1):
    plt.subplot(2, 4, i)
    g = sns.barplot(x=f"{column}", y='Attrition', data=veri)
    g.set_xticklabels(g.get_xticklabels(), rotation=90)
    plt.ylabel('Attrition Count')
    plt.xlabel(f'{column}')
```

```python
#Data Processing
categorical_col.remove('Attrition')
```

```python
# Transform categorical data into dummies
# categorical_col.remove("Attrition")
# data = pd.get_dummies(df, columns=categorical_col)
# data.info()
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
for column in categorical_col:
    veri[column] = label.fit_transform(veri[column])
```

```python
from sklearn.model_selection import train_test_split

X = veri.drop('Attrition', axis=1)
y = veri.Attrition

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Applying machine learning algorithms

from sklearn.metrics import accuracy_score, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        print(f" Eğitim seti: {accuracy_score(y_train, pred) * 100:.2f}%")
        
    elif train==False:
        pred = clf.predict(X_test)      
        print(f" Test seti: {accuracy_score(y_test, pred) * 100:.2f}%")
```

```python
def feature_imp(veri, model):
    fi = pd.DataFrame()
    fi["feature"] = veri.columns
    fi["importance"] = model.feature_importances_
    return fi.sort_values(by="importance", ascending=False)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(solver='liblinear', penalty='l1')
log_reg.fit(X_train, y_train)

print_score(log_reg, X_train, y_train, X_test, y_test, train=True)
print_score(log_reg, X_train, y_train, X_test, y_test, train=False)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(random_state=42, max_depth=5)
tree_clf.fit(X_train, y_train)

print_score(tree_clf, X_train, y_train, X_test, y_test, train=True)
print_score(tree_clf, X_train, y_train, X_test, y_test, train=False)
```

```python
df = feature_imp(X, tree_clf)[:40]
df.set_index('feature', inplace=True)
df.plot(kind='barh', figsize=(7, 7))
plt.title('Feature Importance according to decision trees')
```

```python
from sklearn.ensemble import RandomForestClassifier

rand_forest = RandomForestClassifier(n_estimators=1200)
rand_forest.fit(X_train, y_train)

print_score(rand_forest, X_train, y_train, X_test, y_test, train=True)
print_score(rand_forest, X_train, y_train, X_test, y_test, train=False)
```

```python
df = feature_imp(X, rand_forest)[:40]
df.set_index('feature', inplace=True)
df.plot(kind='barh', figsize=(7, 7))
plt.title('Feature Importance according to Random Forest')
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

print_score(knn, X_train, y_train, X_test, y_test, train=True)
print_score(knn, X_train, y_train, X_test, y_test, train=False)
```
